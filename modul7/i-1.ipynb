# Opgave 2

Hvad gør følgende kode?

```python
x = 11
print(x // 4)
print(x % 4)
print(x / 4)
```


````python
x = 11          # Sets x to the value 11
print(x // 4)   # Will give 2 as this is divisor without floats.
print(x % 4)    # Will give 3, as it's the remainder 
print(x / 4)    # Will give 2.75, as we are working with floats now.
````
